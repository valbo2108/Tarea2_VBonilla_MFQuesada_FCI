```python
# -*- coding: utf-8 -*-
"""
Created on Wends Mar 23 9:38:25 2022

@authors: Cris & Jonathan
"""
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spint
from scipy.integrate import solve_ivp
import pandas as pds

def F(t,v):                    #Se define la función F con los parámetros "t" y "v".
    
    """Esta función calcula el valor de la función F que corresponde a la EDO de primer orden y'=F(t, v)
    Parámetros de la función
    ------------------------
    v : velocidad para un punto t particular
    t : valor particular de t
    p : densidad del material del objeto
    pf: densidad del fluido
    g : aceleración de la gravedad
    a : viscosidad del fluido
    
    Salida de la función
    --------------------
    valorF : valor de la función evaludada
    """
    valorF = ((p-pf)/p)*g-a*v   #Se define la EDO dada.
    return valorF

t = np.linspace(0, 1, 20)       #Se utiliza mediante numpy la función de linspace.
                                #Se le da el valor de inicio, final y un total de 20 puntos para cumplir el espaciado de 0.05s.
v0 = 0                          #Se escribe la condición inicial.
                                #Se escriben las variables con los valores dados.
p = 7.874
pf = 1.260
g = 9.8
a = 10
sol = spint.solve_ivp(F, [0, 1], [v0], t_eval=t, method='RK45')     #Se utiliza mediante scipy la función de solve_ivp, aplicada para el método de Runge Kutta 45.
                                                                    #La función solve_ivp retorna un "bunch", por lo que se utiliza "sol.y".
                                                                    #Para acceder con el nombre de la propiedad contenida dentro del "bunch".
datos= sol.y[0]                                                     #El ".y" corresponde a "Values of the solution at t".
                                                                    #Se utiliza la biblioteca pandas para mostrar los datos obtenidos mediante una tabla.
pds.DataFrame(data={'tiempo':t,'velocidad':datos}, index=None, columns= None, copy= None)
```
